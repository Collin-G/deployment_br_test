```python
import imageio

def get_total_frames_imageio(video_file):
    reader = imageio.get_reader(video_file)
    total_frames = reader.count_frames()
    reader.close()  # Clean up resources
    return total_frames
```

```python
import dlib
import numpy as np

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Helper function to extract facial landmarks
def get_landmarks(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)
    if len(faces) == 0:
        return None  # No face detected

    landmarks = predictor(gray, faces[0])  # Assuming 1 face per frame
    points = np.array([[p.x, p.y] for p in landmarks.parts()], dtype=np.float32)
    return points

# Process video and print landmark points for each frame
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)
    total_frames = get_total_frames_imageio(video_path)
    print("Total frames:", total_frames)

    frame_count = 0  # To count processed frames

    while True:
        ret, frame = cap.read()
        if not ret:
            break  # End of video

        landmarks = get_landmarks(frame)
        frame_count += 1  # Increment the frame count
        # points = np.array([[p.x, p.y] for p in landmarks.parts()], dtype=np.float32)

        if landmarks is not None:
            # print(f"Frame {frame_count}: {landmarks}")  # Print landmarks for the current frame
            print(landmarks.shape)
        else:
            # print(f"Frame {frame_count}: No landmarks detected")
            print("no")

    cap.release()
    cv2.destroyAllWindows()
```
